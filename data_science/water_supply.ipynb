```python
def coords(row):
    add = row['Project Name']
    apiURL = "https://maps.googleapis.com/maps/api/geocode/json?address="+urllib.parse.quote_plus(add+", Los Angeles")+"&key=AIzaSyC4v9aI9v7_MSNOnxDEoIzyVd8XCgGfKP4"
    with urllib.request.urlopen(apiURL) as url:
        try:
            data = json.loads(url.read().decode())['results'][0] #results json from api call
            coords_dict = data['geometry']['location'] #latitude, longitude data
        except:
            coords_dict = {'lat':0, 'lng':0}
    return coords_dict
```
