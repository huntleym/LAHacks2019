# Populating latitude and longitude using Google Maps API

```python
# using approx_address column, calculate 
def lat(row):
    add = row['approx_address']
    apiURL = "https://maps.googleapis.com/maps/api/geocode/json?address="+urllib.parse.quote_plus(add)+"&key=AIzaSyC4v9aI9v7_MSNOnxDEoIzyVd8XCgGfKP4"
    with urllib.request.urlopen(apiURL) as url:
        data = json.loads(url.read().decode())['results'][0] #results json from api call
        lat = data['geometry']['location']['lat'] #latitude, longitude data
    return lat

def lng(row):
    add = row['approx_address']
    apiURL = "https://maps.googleapis.com/maps/api/geocode/json?address="+urllib.parse.quote_plus(add)+"&key=AIzaSyC4v9aI9v7_MSNOnxDEoIzyVd8XCgGfKP4"
    with urllib.request.urlopen(apiURL) as url:
        data = json.loads(url.read().decode())['results'][0] #results json from api call
        lng = data['geometry']['location']['lng'] #latitude, longitude data
    return lng

# x=dtemp.apply(coords, axis=1)
data['lat'] = data.apply(lat, axis=1)
data['lng'] = data.apply(lng, axis=1)
```
